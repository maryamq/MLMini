## TSNE ##
T-SNE reduces dimensionality while trying to keep similar instances close and dissimilar instances apart. It is mostly used for visualization e.g. clusters, data in high dimensional spaces such as images. 
This notebook is a simple example to visualize MNIST data via T-SNE. 

```python
import numpy as np
from sklearn.manifold import TSNE
from sklearn.datasets import fetch_mldata
```

```python
# Load data
mnist = fetch_mldata("MNIST original")
X = mnist["data"]
y = mnist["target"]

num_samples = 1000

# Sample some data. Otherwise it will take forever to fit/transform.
# Also randomize to get different labels.
idx = np.random.randint(len(X), size=1000)
X_train = X[idx]
y_train = y[idx]
```

```python
# Embed data using TSNE.
X_embedded = TSNE(n_components=2).fit_transform(X_train)
```

```python
# To plot pretty figures
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt
plt.rcParams['axes.labelsize'] = 14
plt.rcParams['xtick.labelsize'] = 12
plt.rcParams['ytick.labelsize'] = 12

# Colors for 10 labels.
colors = ['ys', 'ms', 'cs', 'rs','gs', 'bs', 'yo', 'bo', 'mo', 'co' ]

for num in range(0, 10):
    plt.plot(X_embedded[y_train==num, 0], X_embedded[y_train==num, 1], colors[num], label=num)
plt.legend(loc="best", fontsize=10)
plt.show()
```
